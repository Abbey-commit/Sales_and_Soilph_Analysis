```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  

```

```python
# Create a sample dataset  
data = {  
    'Date': pd.date_range(start='2023-01-01', periods=100, freq='D'),  
    'Sales': np.random.randint(100, 1000, size=100),  # Random sales data between 100 and 1000  
    'Category': np.random.choice(['Electronics', 'Clothing', 'Home', 'Toys'], size=100)  # Random categories  
}  
```

```python
# Create a DataFrame  
df = pd.DataFrame(data)

# Display the first few rows of the dataset  
print("Sample Dataset:")  
print(df.head())  
```

```python
# Data Analysis  
# Total sales  
total_sales = df['Sales'].sum()  
print(f"\nTotal Sales: ${total_sales}")  
```

```python
# Average sales  
average_sales = df['Sales'].mean()  
print(f"Average Sales: ${average_sales:.2f}")  
```

```python
# Sales by category  
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)  
print("\nSales by Category:")  
print(sales_by_category)  
```

```python
# Visualization  
plt.figure(figsize=(14, 6))  

# Total Sales Over Time  
plt.subplot(1, 2, 1)  
plt.plot(df['Date'], df['Sales'], color='blue', marker='o', linestyle='-', markersize=4)  
plt.title('Daily Sales Over Time')  
plt.xlabel('Date')  
plt.ylabel('Sales Amount')  
plt.xticks(rotation=45)  
plt.grid()  

# Sales by Category  
plt.subplot(1, 2, 2)  
sales_by_category.plot(kind='bar', color='orange')  
plt.title('Total Sales by Category')  
plt.xlabel('Category')  
plt.ylabel('Total Sales')  
plt.xticks(rotation=45)  
plt.grid()  

plt.tight_layout()  
plt.show()  
```
